```python
from pyspark.sql.functions import *
```

```python
%sql
CREATE VOLUME airbnb1.bronze.bronze_volume;
```

```python
df = spark.read.format("csv")\
               .option("header", "true")\
               .option("inferSchema", "true")\
               .option("multiline","true")\
               .option("escape",'"')\
               .option("quote",'"')\
               .load("/Volumes/databricks_airbnb_sample_data/v01/sf-listings/sf-airbnb.csv")
```

```python
display(df)
```

```python
df.write.format("delta")\
        .mode("append")\
        .saveAsTable("airbnb1.bronze.bronze_data")
```
